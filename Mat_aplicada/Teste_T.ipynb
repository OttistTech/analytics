## <center> Teste T do bot√£o da despensa

Nesta an√°lise checamos se o bot√£o para acesso da despensa acima dos Big Numbers (bem na home do app) afeta a velocidade e intuitivdade do aplicativo na tarefa. Nossa hip√≥tese 1 √© de que o tempo do teste B seja menor, afinal, tentamos colocar sua localiza√ß√£o de forma mais clara para o usu√°rio:

H0 = xB >= xA

H1 = xB < xA

```python
numerador_despensa_t = media_tempo_despensa_B - media_tempo_despensa_A
denominador_despensa_t = math.sqrt((variancia_tempo_despensa_A / 30) + (variancia_tempo_despensa_A / 30))
t_despensa = numerador_despensa_t / denominador_despensa_t
print(t_despensa)
```

```python
numerador_despensa_df = ((variancia_tempo_despensa_A/30) + (variancia_tempo_despensa_B/30))**2
denominador_despensa_df = ((variancia_tempo_despensa_A/30)**2/(30-1)) + ((variancia_tempo_despensa_B/30)**2/(30-1))
df_despensa = numerador_despensa_df / denominador_despensa_df
print(df_despensa)
```

```python
alpha = 0.05
t_critical = t.ppf(1 - alpha, df_despensa)

x = np.linspace(-4, 4, 1000)
y = t.pdf(x, df_despensa)

plt.plot(x, y, label="Distribui√ß√£o t-Student", color="blue")
plt.fill_between(x, 0, y, where=(x >= t_critical), color="red", alpha=0.5, label="Regi√£o de Rejei√ß√£o (One-tailed)")

plt.axvline(t_critical, color='red', linestyle="--", label=f"Valor cr√≠tico (Œ±=0.05): {t_critical:.2f}")
plt.axvline(t_despensa, color='green', linestyle="--", label=f"t calculado: {t_despensa:.2f}")

plt.xlabel("Valores de t")
plt.ylabel("Densidade de Probabilidade")
plt.title(f"Distribui√ß√£o t-Student One-tailed (df = {df_despensa:.2f})")
plt.legend()
plt.show()
```

Como podemos ver, n√£o descartamos a hipot√©se 0 de que o bot√£o acima realmente afete no fator de velocidade do usu√°rio ou intuitividade do acesso. Ou seja, n√£o descartamos que xB >= xA. Iremos encontrar outra forma de resolucionar o problema em uma segunda vers√£o.

## <Center> Teste T do bot√£o das receitas

Nesta an√°lise subimos o bot√£o de preparo da receita, para que assim ele chegue mais r√°pido at√© o local onde indica que ele j√° preparou a receita. Ou seja, o teste √© voltado para velocidade do t√©rmino da tarefa.

H0 = xB >= xA

H1 = xB < xA

```python
numerador_receita_t = media_tempo_receita_A - media_tempo_receita_B
denominador_receita_t = math.sqrt((variancia_tempo_receita_A / 30) + (variancia_tempo_receita_A / 30))
t_receita = numerador_receita_t / denominador_receita_t
print(t_receita)
```

```python
numerador_receita_df = ((variancia_tempo_receita_A/30) + (variancia_tempo_receita_B/30))**2
denominador_receita_df = ((variancia_tempo_receita_A/30)**2/(30-1)) + ((variancia_tempo_receita_B/30)**2/(30-1))
df_receita = numerador_receita_df / denominador_receita_df
print(df_receita)
```

```python
alpha = 0.05
t_critical = t.ppf(1 - alpha, df_receita)

x = np.linspace(-4, 4, 1000)
y = t.pdf(x, df_receita)

plt.plot(x, y, label="Distribui√ß√£o t-Student", color="blue")
plt.fill_between(x, 0, y, where=(x >= t_critical), color="red", alpha=0.5, label="Regi√£o de Rejei√ß√£o (One-tailed)")

plt.axvline(t_critical, color='red', linestyle="--", label=f"Valor cr√≠tico (Œ±=0.05): {t_critical:.2f}")
plt.axvline(t_receita, color='green', linestyle="--", label=f"t calculado: {t_receita:.2f}")

plt.xlabel("Valores de t")
plt.ylabel("Densidade de Probabilidade")
plt.title(f"Distribui√ß√£o t-Student One-tailed (df = {df_receita:.2f})")
plt.legend()
plt.show()
```

Aqui n√≥s descartamos o H0, ou seja, o fato de termos colocado o bot√£o ao in√≠cio da receita, ajudou o usu√°rio a entender onde fica a parte interativa da receita. Mesmo assim, observamos um fator de melhora e "deixamos no radar" uma poss√≠vel mudan√ßa no que est√° escrito neste bot√£o.

```python
alpha = 0.05
t_critical = t.ppf(1 - alpha, 40)

x = np.linspace(-4, 4, 1000)
y = t.pdf(x, 40)

plt.plot(x, y, label="Distribui√ß√£o t-Student", color="blue")
plt.fill_between(x, 0, y, where=(x >= t_critical), color="red", alpha=0.5, label="Regi√£o de Rejei√ß√£o (One-tailed)")

plt.axvline(t_critical, color='red', linestyle="--", label=f"Valor cr√≠tico (Œ±=0.05): {t_critical:.2f}")
plt.axvline(t_receita, color='green', linestyle="--", label=f"t calculado: {t_receita:.2f}")

plt.xlabel("Valores de t")
plt.ylabel("Densidade de Probabilidade")
plt.title(f"Distribui√ß√£o t-Student One-tailed (df = {40:.2f})")
plt.legend()
plt.show()
```

Este √© um teste a mais proposto pelo Daniel. Tentamos tornar o teste mais rigoroso e colocamos o valor 40, aumentando o valor da v√°riavel t calculado. Mesmo assim, a hipot√©se H0 √© rejeitada.
