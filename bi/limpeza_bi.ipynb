## <center> Tratamento de dados de BI </center>

#### Importações

```python
import pandas as pd
```

#### Leitura do DF

```python
df = pd.read_excel('Pesquisa sobre Apps de Despensa e Receitas(1-90).xlsx')
display(df.head(5))
```

#### Excluindo colunas irrelevantes

```python
df = df.drop(['Hora de início', 'Hora de conclusão', 'Email', 'Nome', 'Hora da última modificação'], axis=1)
```

#### Simplificando a nomenclatura das colunas

```python
df.columns = df.columns.str.strip().str.lower()

df = df.rename(columns={
    'qual sua idade?': 'Idade',
    'você mora com sua família?': 'MoraComFamilia',
    'você compra ou compraria alimentos mais próximos da data de validade por preços mais baixos?': 'CompraAlimentosProxValidade',
    'você encontra algum tipo de dificuldade para fazer receitas no cotidiano?': 'DificuldadeFazerReceitas',
    'você utiliza algum aplicativo para buscar receitas?': 'UsaAppParaReceitas',
    'se sim, qual?': 'QualAppReceitas',
    'você já utilizou alguma ia para te ajudar a fazer receitas?': 'UsaIAParaReceitas',
    'você acha interessante receber indicações de receitas práticas de acordo com o que você tem na despensa da sua casa?': 'InteressadoEmSugestoesReceitas',
    'quais recursos você ou sua família utilizam para montar a lista de compras?': 'RecursosListaCompras',
    'você ou sua família já perderam algum produto alimentício por conta da data de validade?': 'PerdeuAlimentoPorValidade',
    'na sua família, vocês costumam verificar a data de validade dos produtos alimentícios regularmente, ou apenas olham a data de validade quando vão usar um produto para preparar uma receita?': 'VerificaValidadeRegularmente',
    'antes de fazer compras, você faz pesquisas para saber em qual supermercado está mais baratos os produtos que deseja comprar?': 'PesquisaPrecosAntesDeComprar',
    'você utiliza algum aplicativo que compara preços de produtos de diferentes mercados?': 'UsaAppComparacaoPrecos',
    'se sim, qual?2': 'QualAppComparacaoPrecos',
    'você tem o hábito de comprar alimentos online?': 'CompraAlimentosOnline',
    'o que você faz com alimentos perto da data de validade?': 'AcaoComAlimentosProxValidade'
})

display(df)
```

#### Tratando os dados da coluna "RecursosListaCompras" para simplificação das respostas

```python
# Dicionário de palavras-chave
palavras_chave = {
    'whatsapp': 'Celular',
    'celular': 'Celular',
    'google keep': 'Celular',
    'bloco de notas': 'Papel',
    'caderno': 'Papel',
    'lista': 'Papel',
    'papel': 'Papel',
    'agenda': 'Papel',
    'post-it': 'Papel',
    'não': 'Não monta lista',
    'nada': 'Não monta lista',
    'memória': 'Não monta lista',
    'aplicativo': 'App de Compras',
    'alexa': 'Alexa',
}

def simplificar_resposta_recursos(resposta):
    resposta = resposta.lower().strip() 

    for palavra, categoria in palavras_chave.items():
        if palavra in resposta:
            return categoria
    return 'Outro'

df['RecursosListaCompras'] = df['RecursosListaCompras'].apply(simplificar_resposta_recursos)
display(df)
```

#### Tratando os dados da coluna "AcaoComAlimentosProxValidade" para simplificação das respostas

```python
# Dicionário de palavras-chave
palavras_chave = {
    'descarto': 'Descarta',
    'nada': 'Descarta',
    'receita': 'Consumo',
    'utilizo': 'Consumo',
    'utilizamos': 'Consumo',
    'usamos': 'Consumo',
    'consumo': 'Consumo',
    'uso': 'Consumo',
    'comemos': 'Consumo',
    'como': 'Consumo',
    'preparo': 'Consumo',
    'consumir': 'Consumo',
}

def simplificar_resposta_recursos(resposta):
    resposta = resposta.lower().strip() 

    for palavra, categoria in palavras_chave.items():
        if palavra in resposta:
            return categoria
    return 'Outro'

df['AcaoComAlimentosProxValidade'] = df['AcaoComAlimentosProxValidade'].apply(simplificar_resposta_recursos)
display(df)
```

#### Exportando a base limpa para BI

```python
df.to_excel('dados_limpos.xlsx', index=False)
```
